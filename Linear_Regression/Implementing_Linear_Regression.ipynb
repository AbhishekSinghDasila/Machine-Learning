```python
class MyLR:

    def __init__(self):
        self.m=None
        self.b=None

    def fit(self,X_train,Y_train):
        num=0
        den=0

        for i in range(X_train.shape[0]):
            num += ((X_train[i]-X_train.mean())*(Y_train[i]-Y_train.mean()))
            den += (X_train[i]-X_train.mean())*(X_train[i]-X_train.mean())

        self.m=(num/den)
        self.b=Y_train.mean()-(self.m*(X_train.mean()))

        print(f"The value of the slope is {self.m}")
        print(f"The value of the intersept is {self.b}")

    def predict(self,X_test):

        return (self.m)*(X_test) + self.b              ## Y=MX+C
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("cgpa_vs_placement_110_students.csv")
```

```python
df.head(10)
```

```python
X= df.iloc[:,0].values          ## select all rows of column=1
Y=df.iloc[: ,1].values           ## select all rows of column=2

# .values convert into numpy array
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3,random_state=3)
```

```python
X_train.shape
```

```python
Lr=MyLR()
```

```python
Lr.fit(X_train,Y_train)
```

```python
## PREDICTION WITH TEST_SIZE=0.3 AND RANDOM_STATE=3
```

```python
print(Lr.predict(X_test[0]))               ## predicted value                        
```
