```python
## Predict  Server CPU Load from Network Traffic
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
x=np.array([10,30,40,50,60]).reshape(-1,1)
```

```python
y=np.array([15,25,50,67,87])
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred= model.predict(x)
```

```python
mae=mean_absolute_error(y,y_pred)
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y,y_pred)
```

```python
n = len(y)
p = x.shape[1]
adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
```

```python
print("MAE = " ,mae)
```

```python
print("MSE = " ,mse)
```

```python
print("RMSE = " ,rmse)
```

```python
print("r2_score = " ,r2)
```

```python
print("r2_adj = " ,adj_r2)
```
