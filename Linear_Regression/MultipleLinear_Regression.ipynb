```python
import numpy as np
from sklearn.datasets import load_diabetes

X,y=load_diabetes(return_X_y=True)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
class MLR:
    def __init__(self):
        self.coeff=None
        self.intercept=None

    def train(self,X_train,y_train):
        X_train=np.insert(X_train,0,1,axis=1)      ## to add column with all 1

        betas=np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)         ##np.linalg.pinv(X_train.T @ X_train) @ X_train.T @ y_train  since it is much safer does not crash for singular matrix also
        self.intercept=betas[0]      #B0
        self.coeff=betas[1:]

    def predict(self,X_test):
        predicted_value= self.intercept + np.dot(X_test,self.coeff)
        ##  Y= Bo + X1B1 + X2B2 + ------------------- + XnBn
        return predicted_value;
```

```python
lr=MLR()
```

```python
lr.train(X_train,y_train)
```

```python
value=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
r2=r2_score(y_test,value)
```

```python
r2
```

```python
mean_absolute_error(y_test,value)
```

```python
mean_squared_error(y_test,value)
```
