```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.tree import plot_tree
```

```python
iris = datasets.load_iris()

X = iris.data
y = iris.target

feature_names = iris.feature_names
```

```python
model_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=3)
model_entropy.fit(X, y)
```

```python
model_gini = DecisionTreeClassifier(criterion="gini", max_depth=3)
model_gini.fit(X, y)
```

```python
plt.figure(figsize=(12,8))
plot_tree(model_entropy, feature_names=feature_names, class_names=iris.target_names, filled=True)
plt.title("Decision Tree using Entropy")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plot_tree(model_gini, feature_names=feature_names, class_names=iris.target_names, filled=True)
plt.title("Decision Tree using Gini")
plt.show()
```

```python
def entropy(y):
    classes, counts = np.unique(y, return_counts=True)
    probs = counts / counts.sum()
    return -np.sum(probs * np.log2(probs))
```

```python
print("Entropy of full dataset:", entropy(y))
```

```python
def gini(y):
    classes, counts = np.unique(y, return_counts=True)
    probs = counts / counts.sum()
    return 1 - np.sum(probs**2)
```

```python
print("Gini of full dataset:", gini(y))
```

```python
def information_gain(X_column, y, threshold):
    parent_entropy = entropy(y)
    
    left_mask = X_column < threshold
    right_mask = X_column >= threshold
    
    if len(y[left_mask]) == 0 or len(y[right_mask]) == 0:
        return 0
    
    n = len(y)
    n_left = len(y[left_mask])
    n_right = len(y[right_mask])
    
    child_entropy = (n_left/n)*entropy(y[left_mask]) + \
                    (n_right/n)*entropy(y[right_mask])
    
    return parent_entropy - child_entropy
```

```python
threshold = np.mean(X[:,0])
print("Information Gain:", information_gain(X[:,0], y, threshold))
```

```python
p = np.linspace(0.001, 0.999, 100)

entropy_vals = -(p*np.log2(p) + (1-p)*np.log2(1-p))
gini_vals = 2*p*(1-p)

plt.plot(p, entropy_vals)
plt.title("Entropy Curve")
plt.xlabel("Probability")
plt.ylabel("Entropy")
plt.show()

plt.plot(p, gini_vals)
plt.title("Gini Curve")
plt.xlabel("Probability")
plt.ylabel("Gini")
plt.show()
```
