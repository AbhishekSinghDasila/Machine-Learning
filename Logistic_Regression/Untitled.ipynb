```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

plt.rcParams["animation.html"] = "jshtml"

# ---------------------------
# 1. Create dataset
# ---------------------------
np.random.seed(2)

class1 = np.random.randn(20,2) + np.array([2,2])
class2 = np.random.randn(20,2) + np.array([-2,-2])

X = np.vstack((class1,class2))
y = np.hstack((np.ones(20), -np.ones(20)))   # +1 and -1 labels


# ---------------------------
# 2. Train perceptron
# ---------------------------
w = np.zeros(2)
b = 0

m_history = []
b_history = []

epochs = 40

for epoch in range(epochs):
    for i in range(len(X)):
        if y[i]*(np.dot(w,X[i]) + b) <= 0:
            w = w + y[i]*X[i]
            b = b + y[i]

    # store line after each epoch
    if w[1] != 0:   # avoid division by zero
        slope = -w[0]/w[1]
        intercept = -b/w[1]
    else:
        slope = 0
        intercept = 0

    m_history.append(slope)
    b_history.append(intercept)


# ---------------------------
# 3. Plot + Animation
# ---------------------------
fig, ax = plt.subplots(figsize=(8,5))

ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=80)

x_i = np.linspace(-6,6,100)
line, = ax.plot(x_i, x_i*m_history[0] + b_history[0],'r-',linewidth=2)

ax.set_ylim(-6,6)
ax.set_xlim(-6,6)

def update(i):
    line.set_ydata(x_i*m_history[i] + b_history[i])
    ax.set_title(f"Epoch {i+1}")
    return line,

anim = FuncAnimation(fig, update, frames=len(m_history), interval=300)

HTML(anim.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=2)

# Add bias column
X_b = np.c_[np.ones((X.shape[0],1)), X]

# Sigmoid
def sigmoid(z):
    return 1/(1+np.exp(-z))

# Training
lr = 0.1
epochs = 100
w = np.zeros(X_b.shape[1])

history = []

for i in range(epochs):
    z = X_b.dot(w)
    y_hat = sigmoid(z)
    
    gradient = X_b.T.dot(y_hat - y)/len(y)
    w -= lr * gradient
    
    if i % 10 == 0:
        history.append(w.copy())

# Plot
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=80)

x_vals = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)

for w_i in history:
    slope = -w_i[1]/w_i[2]
    intercept = -w_i[0]/w_i[2]
    y_vals = slope*x_vals + intercept
    plt.plot(x_vals, y_vals, alpha=0.3)

# Final boundary
slope = -w[1]/w[2]
intercept = -w[0]/w[2]
plt.plot(x_vals, slope*x_vals+intercept, color='red', linewidth=3, label='Final boundary')

plt.legend()
plt.title("How Logistic Regression Learns Decision Boundary")
plt.show()
```
